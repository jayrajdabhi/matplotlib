```python
x = np.linspace(0,10,100)
y = np.sin(x)

```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=((1,1)))
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=((5,5)))
plt.scatter(x,y,color='black', marker='d', s=9)
plt.show()
```

```python
jayraj = plt.figure(figsize=(5,3))
plt.plot(x,y,label='sin(x)', linestyle='--')
plt.plot(x, np.cos(x), label='cos(x)', marker='d')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()


```

```python
plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y)

plt.subplot(2,2,2)
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)
plt.plot(x, np.tanh(x))

plt.subplot(2,2,4)
plt.plot(x, np.tan(x))
plt.show()
```

```python
fig, ax= plt.subplots(2,2)
ax[0,0].plot(x,np.cos(x))
ax[0,1].plot(x,np.tan(x))
ax[1,0].plot(x,np.tanh(x))
ax[1,1].plot(x,y)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.show()
```
